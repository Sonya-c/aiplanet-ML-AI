# Introduction to Data Visualization with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
sma = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/Standard_Metropolitan_Areas_Data-data.csv")
```

```python
sma.head()
```

```python
plt.scatter(sma.crime_rate, sma.percent_senior)
plt.title("crime rate vs percent senior")
plt.xlabel("Percente senior")
plt.ylabel("crime rate")
plt.show()
```

```python
plt.plot(sma.work_force, sma.income)
plt.title("work force vs income")
plt.xlabel("income")
plt.ylabel("work force")
plt.show()
```

```python
plt.hist(sma.percent_senior)
plt.title("percent senior")
plt.show()
```

```python
plt.bar(sma.region, sma.crime_rate)
plt.title("crime rate")
plt.show()
```

```python
plt.barh(sma.region, sma.crime_rate)
plt.title("region vs crime rate")
plt.show()
```

```python
plt.pie(sma.region.value_counts(), explode=[.1, 0, 0, 0], labels=sma.region.unique(), autopct="%1.1f%%")
plt.title("Region")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(sma.work_force, sma.income, label="work force", color="blue", linestyle="--", marker="o")
plt.plot(sma.physicians, sma.income, label="physicians", color="c", linestyle="-.", marker="x")
plt.plot(sma.hospital_beds, sma.income, label="hospital beds", color="0.3", linestyle=":", marker="^")

plt.title("work force & physucians & hospital beds vs income")
plt.xlabel("income")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(sma.work_force, sma.income, label="work force", color="blue", linestyle="--", marker="o")
plt.title("work force vs income")

plt.subplot(1, 2, 2)
plt.plot(sma.physicians, sma.income, label="physicians", color="c", linestyle="-.", marker="x")
plt.title("physucians vs income")


plt.xlabel("income")
plt.legend()

plt.suptitle("work force & physicians")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(sma.work_force, sma.income, label="work force", color="blue", linestyle="--", marker="o")
ax1.set_title("work force vs income")

ax2.plot(sma.physicians, sma.income, label="physicians", color="c", linestyle="-.", marker="x")
ax2.set_title("physucians vs income")

plt.suptitle("work force & physicians")
plt.show()
```
